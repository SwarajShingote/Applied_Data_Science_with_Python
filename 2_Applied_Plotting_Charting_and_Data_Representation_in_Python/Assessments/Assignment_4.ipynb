```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib


channel=["Aaj Tak", "ABP News", "India Tv", "Zee News","NDTV"]
TV_impression=[893, 635, 532, 343,252]


ax = sns.barplot(channel,TV_impression)

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center', xytext = (0, 10),
                textcoords = 'offset points')


ax.set(ylim=(0, 1000))

plt.title("Most Viewed News Channel in India",fontsize =20)
plt.xlabel("News Channels",fontsize =10)
plt.ylabel("TV impressions (in thousands)",fontsize =10)

plt.xticks(fontsize =10)
plt.yticks(fontsize =10)

name_to_color = {
    'Aaj Tak':   'blue',
    'ABP News':   'orange',
    'India TV': 'green',
    'Zee News': 'red',
    'NDTV' : 'purple'
}

patches = [matplotlib.patches.Patch(color=v, label=k) for k,v in name_to_color.items()]
matplotlib.pyplot.legend(handles=patches)


plt.tight_layout();
plt.show();
```
